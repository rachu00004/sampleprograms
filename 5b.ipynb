auto pct displays the percnetage upto 1 decimal points

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
X=[34,89,12,78]
```

```python
plt.figure(figsize=(5,5))
plt.pie(X,autopct='%.2f%%',radius=1.2,labels=['sat','sun','mon','tues'])
plt.show()
```
